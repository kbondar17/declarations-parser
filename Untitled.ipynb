```python
import requests
import time
import os
import urllib.request
import pandas as pd
%config Completer.use_jedi = False
from pathlib import Path
# import tqdm.notebook.tqdm as tqdm
from tqdm import tqdm_notebook
# from tqdm.notebook import tqdm
df = pd.read_csv('salaries_with_ext.csv')
df.head()
df['extension'].unique()
```

```python
# # 580it

for k,v in tqdm_notebook(df.iterrows(), total=len(df)):
    id_ = v['document_file_id']

    folder = Path(f"data_ids/{v['extension']}")
    
    if not folder.exists():
        folder.mkdir()

 
    filename=v['file'].split('filename="')[-1].strip('"')
    filename = str(id_) + '_' + filename
    
    urllib.request.urlretrieve(url=v['link'], filename=folder/filename)
```
